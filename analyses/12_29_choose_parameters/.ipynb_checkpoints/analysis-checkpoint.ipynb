```r
library(dplyr)
library(data.table)
library(tidyr)
library(ggplot2)
```

```r
df = fread(cmd="cat ./data.txt | sed 's/,//g' ") %>%
    separate(col = file, into = c("A","B","C","keepPositionsFraction","topNperchaving_sequences_to_remove"),sep = "_|\\.fa",convert=T)
df$A <- NULL
df$B <- NULL
df$C <- NULL
df
```

```r
options(repr.plot.width=16, repr.plot.height=12)
ggplot(data=df) + geom_point(aes(x=num_seqs,y=avg_len,color=factor(keepPositionsFraction)),size=8) +
theme(text = element_text(size=20),
        axis.text.x = element_text(angle=0)) +
labs(color="Fraction of positions to keep") +
xlab("Number of sequences") +
ylab("Length of sequences")
```
